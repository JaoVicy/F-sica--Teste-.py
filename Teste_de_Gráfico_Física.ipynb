```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros da simulao
A = 100  # Amplitude da acelerao (m/s)
omega = 10 * np.pi  # Frequncia angular (rad/s)
t_max = 10  # Tempo mximo da simulao (s)
dt = 0.01  # Intervalo de tempo (s)

# Vetor de tempo
t = np.arange(0, t_max, dt);

# Funo de acelerao variando com o tempo
a = A * np.sin(omega * t)

# Integrao numrica para encontrar a velocidade e a posio
v = np.cumsum(a) * dt  # Velocidade como a integral de a(t)
x = np.cumsum(v) * dt  # Posio como a integral de v(t)

# Plot dos grficos
plt.figure(figsize=(10, 8))

# Grfico da acelerao
plt.subplot(3, 1, 1)
plt.plot(t, a, label='Acelerao (a)', color='blue')
plt.ylabel('Acelerao (m/s)')
plt.legend(loc="upper right")
plt.grid(True)

# Grfico da velocidade
plt.subplot(3, 1, 2)
plt.plot(t, v, label='Velocidade (v)', color='green')
plt.ylabel('Velocidade (m/s)')
plt.legend(loc="upper right")
plt.grid(True)

# Grfico da posio
plt.subplot(3, 1, 3)
plt.plot(t, x, label='Posio (x)', color='red')
plt.xlabel('Tempo (s)')
plt.ylabel('Posio (m)')
plt.legend(loc="upper right")
plt.grid(True)

plt.tight_layout()
plt.show()
```
